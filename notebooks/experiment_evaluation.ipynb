```python
df_model_address = pd.DataFrame([


{'checkpoint':'/home/parham/entity-ranking/src/lightning_logs/version_5/checkpoints/epoch=0-step=18418.ckpt'},
{'checkpoint':'/home/parham/entity-ranking/src/lightning_logs/version_6/checkpoints/epoch=0-step=18418.ckpt'},
{'checkpoint':'/home/parham/entity-ranking/src/lightning_logs/version_7/checkpoints/epoch=0-step=18418.ckpt'},
{'checkpoint':'src/lightning_logs/version_8/checkpoints/epoch=0-step=18418.ckpt'},


])
```

```python
df_experiments = df_model_address.merge(databins, how = 'cross')
```

```python
df_experiments
```

```python
## Create bash script commands for score generator script
```

```python
commands = []
for _,row in tqdm(df_experiments.iterrows()):
    command = ["python","score_generator.py" ,
"--table_address",
"{}".format(row.table_address),
"--model_name",
"{}".format(row.checkpoint)]
    commands.append(' '.join(command))
```

```python
## save bash script commands for score generator script
with open('../src/score_generation_script.sh', 'w') as filehandle:
    for listitem in commands:
        filehandle.write('%s\n' % listitem)
```
